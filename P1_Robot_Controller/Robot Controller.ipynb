```
environment = [[0,0,0,2], 
               [1,2,0,0],
               [0,2,3,2]]
```

- 0: 普通格子（可通行）
- 1: 机器人的起点（可通行）
- 2: 障碍物（不可通行）
- 3: 宝藏箱（目标点）
